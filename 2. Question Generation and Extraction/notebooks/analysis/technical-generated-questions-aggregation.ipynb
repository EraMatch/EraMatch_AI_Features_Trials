## Combining Essay Questions

```python
# Essay Question Aggregation
essay_dataframes = [
    df_ml_ds_essay_eval,
    df_dsa_applied_essay_eval,
    df_dsa_abstract_essay_eval,
    df_swe_abstract_essay_eval
]

# 2. Concatenate them into a single DataFrame
df_all_essays = pd.concat(essay_dataframes, ignore_index=True)

# 3. Verify the combined DataFrame
print(f"Combined DataFrame Shape: {df_all_essays.shape}")
```

```python
# Add a source column to identify the domain of the questions
df_ml_ds_essay_eval['source_domain'] = 'ML_DS_Abstract_Theory'
df_dsa_applied_essay_eval['source_domain'] = 'DSA_Applied_Engineering'
df_dsa_abstract_essay_eval['source_domain'] = 'DSA_Abstract_Theory'
df_swe_abstract_essay_eval['source_domain'] = 'SWE_Abstract_Theory'

# Concatenate with the source tracking included
df_all_essays = pd.concat([
    df_ml_ds_essay_eval,
    df_dsa_applied_essay_eval,
    df_dsa_abstract_essay_eval,
    df_swe_abstract_essay_eval
], ignore_index=True)

# Inspect the distribution to confirm it worked
print(df_all_essays['source_domain'].value_counts())
```

```python
import pandas as pd

print("--- Essay Dataset Structure Overview ---")
print(df_all_essays.info())

print("\n--- Percentage of Missing Values ---")
print(df_all_essays.isnull().mean() * 100)

print("\n--- Diagnostic Sample Row ---")
sample_row = df_all_essays.dropna(how='all').head(1)
for col in sample_row.columns:
    print(f"[{col}]: {sample_row[col].values[0]}")
```

```python
import pandas as pd
import numpy as np

df_essay_processed = df_all_essays.copy()

print("--- 1. Initial Data Quality Filtering ---")

initial_count = len(df_essay_processed)
core_cols = [col for col in ['question', 'answer', 'response'] if col in df_essay_processed.columns]
if core_cols:
    df_essay_processed.dropna(subset=core_cols, inplace=True)
    print(f"Dropped {initial_count - len(df_essay_processed)} rows due to missing critical text features.")
else:
    df_essay_processed.dropna(how='all', inplace=True)

print("\n--- 2. Standardizing String Format ---")

def safe_strip(x):
    if isinstance(x, (str, bytes)):
        return x.strip()
    return x

for col in df_essay_processed.columns:
    if df_essay_processed[col].dtype == 'object':
        df_essay_processed[col] = df_essay_processed[col].apply(safe_strip)

print("\n--- 3. Executing Deep Text Length Analysis ---")

text_analytics_cols = [col for col in ['question', 'answer', 'response', 'context'] if col in df_essay_processed.columns]
for col in text_analytics_cols:
    word_counts = df_essay_processed[col].apply(lambda x: len(str(x).split()) if isinstance(x, str) else 0)
    print(f"Column [{col}]:")
    print(f"  Average word count: {word_counts.mean():.1f}")
    print(f"  Longest text (max words): {word_counts.max()}")
    print(f"  Shortest text (min words): {word_counts.min()}")

print("\n--- 4. Engineering Consolidated Tag Features ---")

def combine_essay_tags(row):
    tags = []
    for i in range(3):
        col_name = f'tags/{i}'
        if col_name in row and pd.notna(row[col_name]):
            val = row[col_name]
            if isinstance(val, str):
                tag_str = val.strip()
                if tag_str.lower() != 'nan' and tag_str != '':
                    tags.append(tag_str)
    return tags if tags else np.nan

if 'tags/0' in df_essay_processed.columns:
    df_essay_processed['all_tags'] = df_essay_processed.apply(combine_essay_tags, axis=1)
    df_essay_processed.drop(columns=['tags/0', 'tags/1', 'tags/2'], errors='ignore', inplace=True)

df_essay_processed.reset_index(drop=True, inplace=True)

print("\n==================================================================")
print("WORKFLOW COMPLETE: ESSAY DATASET SYNCHRONIZED AND MODEL-READY")
print(f"Final Dataframe Shape: {df_essay_processed.shape}")
print("==================================================================")

inspection_cols = [col for col in ['question', 'answer', 'response', 'source_domain'] if col in df_essay_processed.columns]
if len(inspection_cols) >= 2:
    print("\nSample Inspection of Verified Dataset Rows:")
    print(df_essay_processed[inspection_cols].sample(min(3, len(df_essay_processed))))
```

## Combining MCQ Questions

```python
# 1. Add the source column to each MCQ DataFrame to track its origin
df_ml_ds_mcq['source_domain'] = 'ML_DS_Abstract_Theory'
df_dsa_applied_mcq['source_domain'] = 'DSA_Applied_Engineering'
df_dsa_abstract_mcq['source_domain'] = 'DSA_Abstract_Theory'
df_swe_abstract_mcq['source_domain'] = 'SWE_Abstract_Theory'

# 2. Gather all the MCQ DataFrames into a list
mcq_dataframes = [
    df_ml_ds_mcq,
    df_dsa_applied_mcq,
    df_dsa_abstract_mcq,
    df_swe_abstract_mcq
]

# 3. Concatenate them into a single DataFrame
df_all_mcqs = pd.concat(mcq_dataframes, ignore_index=True)

# 4. Verify the results
print(f"Combined MCQ DataFrame Shape: {df_all_mcqs.shape}")
print("\nRows per domain:")
print(df_all_mcqs['source_domain'].value_counts())
```

```python
import pandas as pd
import numpy as np

df_processed = df_all_mcqs.copy()

print("--- 1. Initial Data Quality Filtering ---")

df_processed.drop(columns=['distractors/3', 'Field1'], errors='ignore', inplace=True)

initial_count = len(df_processed)
df_processed.dropna(subset=['question', 'choices/0', 'correct_choice', 'extracted_answer'], inplace=True)
print(f"Dropped {initial_count - len(df_processed)} rows due to missing critical question/choice/answer tokens.")

print("\n--- 2. Standardizing String Format ---")

choice_cols = ['choices/0', 'choices/1', 'choices/2', 'choices/3']
for col in choice_cols:
    df_processed[col] = df_processed[col].astype(str).str.strip()

df_processed['correct_choice'] = df_processed['correct_choice'].astype(str).str.strip().str.upper()

print("\n--- 3. Executing Deep Quality Analysis & Smart Alignment ---")

letter_to_col = {'A': 'choices/0', 'B': 'choices/1', 'C': 'choices/2', 'D': 'choices/3'}
col_to_letter = {'choices/0': 'A', 'choices/1': 'B', 'choices/2': 'C', 'choices/3': 'D'}

perfect_matches = 0
repaired_matches = 0
unresolvable_mismatches = 0

def process_and_align_row(row):
    global perfect_matches, repaired_matches, unresolvable_mismatches
    
    current_letter = row['correct_choice']
    extracted_text = str(row['extracted_answer']).lower().strip()
    target_col = letter_to_col.get(current_letter)
    
    clean_text = lambda x: str(x).lower().replace('a)', '').replace('b)', '').replace('c)', '').replace('d)', '').strip()
    
    if target_col and pd.notna(row[target_col]):
        choice_text_cleaned = clean_text(row[target_col])
        if extracted_text in choice_text_cleaned or choice_text_cleaned in extracted_text:
            perfect_matches += 1
            return current_letter, row[target_col]
            
    for col in choice_cols:
        if pd.notna(row[col]):
            alt_choice_cleaned = clean_text(row[col])
            if extracted_text in alt_choice_cleaned:
                repaired_matches += 1
                corrected_letter = col_to_letter[col]
                return corrected_letter, row[col]
                
    unresolvable_mismatches += 1
    fallback_col = target_col if target_col else 'choices/0'
    return current_letter, row[fallback_col]

alignment_results = df_processed.apply(process_and_align_row, axis=1)

df_processed['correct_choice'] = [res[0] for res in alignment_results]
df_processed['verified_correct_text'] = [res[1] for res in alignment_results]

print(f"  Perfectly aligned rows (Letter matched Text): {perfect_matches}")
print(f"  Repaired rows (Letter corrected programmatically): {repaired_matches}")
print(f"  Unresolvable rows (Text absent from choices): {unresolvable_mismatches}")

df_processed.drop(columns=['extracted_answer'], errors='ignore', inplace=True)

print("\n--- 4. Engineering Consolidated Tag Features ---")

def combine_tags(row):
    tags = []
    for i in range(3):
        col_name = f'tags/{i}'
        if col_name in row and pd.notna(row[col_name]):
            tag_str = str(row[col_name]).strip()
            if tag_str.lower() != 'nan' and tag_str != '':
                tags.append(tag_str)
    return tags if tags else np.nan

df_processed['all_tags'] = df_processed.apply(combine_tags, axis=1)
df_processed.drop(columns=['tags/0', 'tags/1', 'tags/2'], errors='ignore', inplace=True)

df_processed.reset_index(drop=True, inplace=True)

print("\n========================================q==========================")
print("WORKFLOW COMPLETE: DATASET SYNCHRONIZED AND MODEL-READY")
print(f"Final Dataframe Shape: {df_processed.shape}")
print("==================================================================")

print("\nSample Inspection of Verified Dataset Rows:")
print(df_processed[['question', 'correct_choice', 'verified_correct_text']].sample(3))
```
